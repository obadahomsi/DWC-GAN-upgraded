```python
max_num_objects = 0
min_num_objects = 10
print(len(clvr_attr['scenes']))
for i in clvr_attr['scenes']:
    if len(i['objects']) > max_num_objects:
        max_num_objects = len(i['objects'])
    if len(i['objects']) < min_num_objects:
        min_num_objects = len(i['objects'])
print(min_num_objects,max_num_objects)
```

```python
unique = []
for i in clvr_attr['scenes']:
    for l in i['objects']:
        if l['material'] not in unique:
            unique.append(l['material'])
    
```
